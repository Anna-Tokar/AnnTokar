```python
# Передача данных с помощью словаря с ключами строками

data={'a':np.arange(50), # [0, 49]
#       Возвращает случайные целые числа (количество - 3 параметр):
      'c':np.random.randint(0,50,50),
#       Возвращает случайную матрицу - количество эл-в 50
      'd':np.random.randn(50)}
data['b']=data['a']+10 * np.random.randn(50)
data['d']=np.abs(data['d'])*100

# создаем диаграмму
plt.scatter('a', 'b', c ='c',s ='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# Работа с несколькими фигурами и осями

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1) # шаг 0.1
t2=np.arange(0.0,5.0,0.02) # шаг 0.02

plt.figure()
plt.subplot(211) # или так: plt.subplot(2, 1, 1)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
# Работа с текстом

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000) # Возвращает случайную матрицу - количество эл-в 10000

# Hist используется для рисования гистограммы, а гистограмма по существу является графиком
# 50 в hist - площадь "коробки" каждой
# Если density = True, то результатом вычислений окажется функция (точнее ее значения) 
# плотности вероятности, а интеграл по диапазону значений range будет равен 1
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,'$\mu=100,\\sigma=15$')
plt.axis([40,160,0,0.03]) # размер окна 
plt.grid(True)  # Настройте линии сетки - включили

t=plt.xlabel('my data',fontsize=14,color='red')

plt.show()
```

```python
# Использование математических выражений в тексте

# для записи выражения σi=15 в заголовке, надо написать TeX-выражение внутри знаков доллара:
plt.title(r'$\sigma_i=15$')
# r, предшествующий строке, важен - сырая строка и не должна обратные косые черты как экранирование python
```

```python
# Аннотации (подписи)

# Функция text помещает текст в произвольное место.
# В annotate задаются 2 точки: расположение аннотируемого объекта xy и расположение текста xytext. 
# Оба этих аргумента задаются кортежами вида(x, y).

ax=plt.subplot()
t=np.arange(0.0,5.0,0.01)  # кривая с шагом 0.01
s=np.cos(2*np.pi*t)

line,=plt.plot(t,s,lw=2)
# arrowprops - словарь, который содержит элементы, задающие внешний вид стрелки.
# shrink - коэффициент, показывающий, какую часть длины указателя будет занимать "голова" стрелки. 
# Например, если shrink = 0.5, то стрелка будет занимать половину длины указателя.
plt.annotate('local max',xy=(2,1),xytext=(3,1.5),arrowprops=dict(facecolor='black',shrink=0.05))
# ylim - Получите или установите y-пределы текущих осей.
plt.ylim(-2,2)
plt.show()
```

```python
# Логарифмические и нелинейные оси координат

# random.seed() задает начальные условия для генератора случайных чисел.
# Генератор инициализированный одними и теми же начальными условиями, выдает абсолютно идентичные случайные последовательности.
np.random.seed(19680801)

# random normal() генерирует массив заданной формы и заполняет его случайными значениями
# loc — указывает среднее (пиковое) значение распределения.
# scale - указывает стандартное отклонение или то, насколько плоским должен быть график распределения. 
# size - указывает форму результирующего массива.
y=np.random.normal(loc=0.5,scale=0.4,size=1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))

plt.figure()  
plt.subplot(221)
plt.plot(x,y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x,y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top=0.92,bottom=0.08,left=0.10,right=0.95,hspace=0.25,  wspace=0.35)

plt.show()
```

```python
# Основы построения графиков:команда plot

# Series — это структура, используемая для работы с последовательностью одномерных данных (типо таблиц)
ts=pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2000",periods=1000))
# cumsum - сумма эл-в ts
ts=ts.cumsum()
ts.plot();
```

```python
# В DataFrame, функция plot() удобна для построения всех столбцов с метками:
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list("ABCD"))
# cumsum - сумма эл-в df
df=df.cumsum()
plt.figure();
df.plot();
```

```python
df3=pd.DataFrame(np.random.randn(1000,2),columns=["B","C"]).cumsum()
df3["A"]=pd.Series(list(range(len(df))))
df3.plot(x="A",y="B");
```

```python
# Другие графики

plt.figure();
# iloc[] - используется для доступа по числовому значению (начиная от 0) — т.е. по номеру строки и номеру столбца
df.iloc[9].plot(kind="bar");
```

```python
# Гистограммы (Histograms)

# Гистограммы можно нарисовать с помощью методов DataFrame.plot.hist() и Series.plot.hist().
df4 = pd.DataFrame(
{
"a": np.random.randn(1000) + 1,
"b": np.random.randn(1000),
"c": np.random.randn(1000) - 1,
},
columns=["a", "b", "c"],
)
plt.figure();
# Hist используется для рисования гистограммы, а гистограмма по существу является графиком
# alpha - прозрачность рисунка
df4.plot.hist(alpha=0.5);
```

```python
# Гистограмму можно сложить в стопку, используя stacked=True. Размер ячейки можно изменить с помощью ключевого слова bins.

plt.figure();
# bins - площади кубов
df4.plot.hist(stacked=True, bins=20);
```

```python
# горизонтальные и кумулятивные гистограммы могут быть нарисованы с помощью orientation='horizontal' and cumulative=True

plt.figure();
df4["a"].plot.hist(orientation="horizontal", cumulative=True);
```

```python
# С помощью by можно задать построение группы гистограмм:

# это объект библиотеки pandas, спроектированный для представления одномерных структур данных, 
# похожих на массивы, но с дополнительными возможностями. 
# Его структура проста, ведь он состоит из двух связанных между собой массивов. 
# Основной содержит данные (данные любого типа NumPy), а в дополнительном, index, хранятся метки.
data = pd.Series(np.random.randn(1000))
data.hist(by=np.random.randint(0, 4, 1000), figsize=(6, 4));
```

```python
# Box plots

# Dataframe — это табличная структура данных, напоминающая таблицы из Microsoft Excel. 
# Ее главная задача — позволить использовать многомерные Series.
df = pd.DataFrame(np.random.rand(10, 5), columns=["A", "B", "C", "D", "E"])
df.plot.box();
```

```python
# Area plot (график с закрашенной областью)

# Чтобы создать график области с накоплением, 
# каждый столбец должен содержать либо все положительные, либо все отрицательные значения.
# NaN автоматически заменяется на 0. Если хотите отбросить или заполнить их другими значениями, 
# используйте dataframe.dropna() or dataframe.fillna() перед вызовом plot

df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.plot.area();
```

```python
# Чтобы создать неупакованный график, передайте stacked=False. 
# Прозрачность (Alpha) устанавлиевается равной 0,5, если не указано иное

df.plot.area(stacked=False);
```
